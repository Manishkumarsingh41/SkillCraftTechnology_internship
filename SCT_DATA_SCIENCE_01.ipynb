```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better plot styling
sns.set(style="whitegrid")
```

```python
# Load CSV into DataFrame
df = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')

# Preview the dataset
df.head()
```

```python
# Gender distribution
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Math score distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['math score'], bins=10, kde=True, color='skyblue')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Parent education level distribution
plt.figure(figsize=(10, 5))
sns.countplot(y='parental level of education', data=df, order=df['parental level of education'].value_counts().index, palette='coolwarm')
plt.title('Parental Level of Education')
plt.xlabel('Count')
plt.ylabel('Parental Education')
plt.show()
```

```python
# Reading score distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['reading score'], bins=10, kde=True, color='orange')
plt.title('Reading Score Distribution')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of math scores by gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='gender', y='math score', data=df, palette='Set2')
plt.title('Math Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```

```python
# Violin plot of reading score by gender
plt.figure(figsize=(8, 5))
sns.violinplot(x='gender', y='reading score', data=df, palette='muted')
plt.title('Reading Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()
```

```python
# Gender pie chart
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['#ffc0cb','#87cefa'])
plt.title('Gender Distribution')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='Blues')
plt.title('Correlation Between Scores')
plt.show()
```

```python
# Pair plot
sns.pairplot(df[['math score', 'reading score', 'writing score']])
plt.suptitle("Score Distributions & Relationships", y=1.02)
plt.show()
```

```python
# Swarm plot of writing scores by gender
plt.figure(figsize=(8, 5))
sns.swarmplot(x='gender', y='writing score', data=df, palette='cool')
plt.title('Writing Score by Gender')
plt.show()
```

```python
# Lollipop chart for average scores by gender
avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

plt.figure(figsize=(10, 5))
for subject in avg_scores.columns:
    plt.plot(avg_scores.index, avg_scores[subject], marker='o', label=subject)

plt.title('Average Scores by Gender (Lollipop Style)')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.legend()
plt.grid(axis='y', linestyle='--')
plt
```

```python
import numpy as np

# Radar chart for Male vs Female average scores
categories = ['Math', 'Reading', 'Writing']
N = len(categories)

# Grouped mean values
male_avg = df[df['gender'] == 'male'][['math score', 'reading score', 'writing score']].mean().values
female_avg = df[df['gender'] == 'female'][['math score', 'reading score', 'writing score']].mean().values

# Repeat first value to close the circle
male_avg = np.concatenate((male_avg, [male_avg[0]]))
female_avg = np.concatenate((female_avg, [female_avg[0]]))
categories += [categories[0]]

angles = np.linspace(0, 2 * np.pi, N + 1)

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(angles, male_avg, 'b-', linewidth=2, label='Male')
ax.fill(angles, male_avg, 'b', alpha=0.1)

ax.plot(angles, female_avg, 'r-', linewidth=2, label='Female')
ax.fill(angles, female_avg, 'r', alpha=0.1)

ax.set_thetagrids(angles * 180/np.pi, categories)
plt.title('Radar Chart: Avg Score by Gender')
plt.legend(loc='upper right')
plt.show()
```

```python
# Faceted histogram: writing score by gender
g = sns.FacetGrid(df, col='gender')
g.map(plt.hist, 'writing score', bins=10, color='skyblue', edgecolor='black')
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('Writing Score Distribution by Gender')
plt.show()
```

```python
df['avg_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df['score_deviation'] = df['avg_score'] - df['avg_score'].mean()
df_sorted = df.sort_values('score_deviation')

colors = ['green' if x < 0 else 'red' for x in df_sorted['score_deviation']]

plt.figure(figsize=(10, 6))
plt.barh(df_sorted.index, df_sorted['score_deviation'], color=colors)
plt.title('Diverging Bar Chart – Score Deviation from Average')
plt.xlabel('Deviation')
plt.ylabel('Student Index')
plt.axvline(0, color='black', linewidth=0.7)
plt.tight_layout()
plt.show()
```
